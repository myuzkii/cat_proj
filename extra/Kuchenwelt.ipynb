```python

######################################################################
# Im Folgenden die Klassen für eine kleine Kuchensimulation als Test #
######################################################################

class Form():
    def __init__(self,radius,hoehe):
        self.radius = radius
        self.hoehe = hoehe

class ChemischesVerhalten():

    def __init__(self,mindesthaltbarkeit,temperatur,bakterienzahl=1,vergoren=False):
        self.mindesthaltbarkeit = mindesthaltbarkeit
        self.temperatur = temperatur
        self.bakterienzahl = bakterienzahl
        self.vergoren = vergoren
        
    def mindesthaltbarkeit_ausgeben(self):
        print(self.mindesthaltbarkeit)    

    def alterung(self):
        self.bakterienzahl = self.bakterienzahl*self.bakterienzahl*self.temperatur
        if self.bakterienzahl > 5000:
            self.vergoren = True
        return self.vergoren
    def get_bakterienzahl(self):
        return self.bakterienzahl  

class Topping():
    def __init__(self,farbe,geschmack):
        self.farbe = farbe
        self.geschmack = geschmack
    def __str__(self):
        return "Farbe: {0}, Geschmack: {1}".format(self.farbe,self.geschmack)


class KaramellTopping(Topping):
    def __init__(self, farbe, geschmack,karamellisierungsgrad = "hoch"):
        super().__init__(farbe, geschmack)
        self.karamellisierungsgrad = karamellisierungsgrad
    def __str__(self):
        return "Farbe: {0}, Geschmack: {1}, Karamellisierungsgrad: {2}".format(self.farbe,self.geschmack,self.karamellisierungsgrad)

class Fuellung():
    def __init__(self,farbe,geschmack):
        self.farbe = farbe
        self.geschmack = geschmack
    def __str__(self):
        return "Farbe: {0}, Geschmack: {1}".format(self.farbe,self.geschmack)

class Kuchen(Form,ChemischesVerhalten):

    def __init__(self,topping,fuellung):
        self.topping = topping
        self.fuellung = fuellung

    def set_form(self,radius,hoehe):           
        self.radius = radius
        self.hoehe = hoehe 

    def set_chemischesVerhalten(self,mindesthaltbarkeit,temperatur):
        self.mindesthaltbarkeit = mindesthaltbarkeit
        self.temperatur = temperatur
        self.bakterienzahl = 1
        self.vergoren = False

    def __str__(self):
        return "Fuellung: [{0}] Topping: [{1}] ".format(self.fuellung,self.topping)
        
########################################################
# Jetzt der Start der Kuchensimulation (Klassenaufruf) #
########################################################

topping1 = KaramellTopping(farbe="braun",geschmack="süß",karamellisierungsgrad="gering")
fuellung1 = Fuellung(farbe="rot",geschmack="sauer")
kuchen1 = Kuchen(topping1,fuellung1)

kuchen1.set_chemischesVerhalten(mindesthaltbarkeit=10,temperatur=5)
kuchen1.set_form(radius=5,hoehe=21)


print(kuchen1)
while True:
    kuchen_zustand_vergoren = kuchen1.alterung()
    if kuchen_zustand_vergoren == True:
        print("Kuchen ist vergoren. Bakterienzahl: ", kuchen1.get_bakterienzahl())
        break
    print("Kuchen ist frisch. Bakterienzahl: ", kuchen1.get_bakterienzahl())    

print("Ende der Kuchensimulation")





```
